```python
import pandas as pd
```

```python
df_adresses = pd.read_csv("../../Data/Addresses.csv")
df_pub = pd.read_csv("../../Data/Publications.csv")
df_pub
```

```python
merged_df = pd.merge(df_pub, df_adresses, on='Institution', how='inner')
merged_df
```

```python
bins = range(2004, 2025, 5)

merged_df['year_bin'] = pd.cut(merged_df['Year'], bins=bins, right=False)

merged_df
```

```python
merged_df.dropna(subset=["year_bin"],inplace=True)
merged_df
```

```python
merged_df["International"] = merged_df["Country"].apply(lambda x: 0 if x == 'Denmark' else 1)

merged_df
```

```python
year_bins = merged_df["year_bin"].unique()
year_bins
```

```python
df_timeframe = []
for bin in year_bins:
    df_timeframe.append(merged_df[merged_df["year_bin"] == bin])
df_timeframe
```

```python
df_final = pd.DataFrame(columns=["Title", "Year"])

for time in df_timeframe:
    help_df = time.sample(25)
    help_df.reset_index(drop=True, inplace=True)
    df_final = pd.concat([df_final, help_df[["Title", "Year"]]], ignore_index=True)


df_final
```

```python
df_final.drop(columns="Year", inplace=True)
```

```python
df_final
```

```python
df_final.to_csv("annotation.csv", index=False)
```
